```python
import numpy as np
import torch

from models.SineNet.SineNet import SineNet
from datasets.meta.SineNetTask.loading_data import loading_data
```

```python
lr = 0.01
```

```python
model = SineNet()
model.make_alpha(lr)  # Handy for comparison to Meta trained model. Not updated in this notebook
model = model.cuda()
```

```python
criterion = torch.nn.MSELoss()
optim = torch.optim.SGD(model.parameters(), lr=lr)
```

```python
train_loader, _ = loading_data()
```

```python
for ep in range(10000):
    loss_sum = 0
    for x_train, y_train, x_test, y_test in train_loader:
        x_train = x_train.squeeze(0).cuda()
        y_train = y_train.squeeze(0).cuda()
        
        optim.zero_grad()
        pred = model.forward(x_train)
        loss = criterion(pred, y_train)
        loss.backward()
        optim.step()
        loss_sum += loss.item()
    print(f'ep {ep + 1}, loss: {loss_sum}')
```

```python
model_state = {}
model_state['net'] = model.state_dict()
torch.save(model_state, 'notebooks/SineNetStates/SineNetNormal_10keps.pth')
```
